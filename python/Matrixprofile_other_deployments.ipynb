```python
def steps_per_sec(dp, window = 1200):
    #dp distance profile
    #window seconds
    x = dp
    for i in range(len(dp)):

        if dp[i] != 0:
            x[i] = 0
        
        else :
            x[i] = 1
    steps_p_s = np.zeros(int(len(dp)/window))
    k=0
    for i in range(0,len(dp),window):
        if k >= len(steps_p_s):
            break
        steps_p_s[k] =  np.sum(x[i:i+window])
        #print(dp[i:i+10])
        #print(steps_p_s[k])
        k+=1
    return steps_p_s

sps_1 = steps_per_sec(dp_15_1)
sps_2 = steps_per_sec(dp_15_2)
sps_3 = steps_per_sec(dp_15_3)
sps_4 = steps_per_sec(dp_15_4)
sps_5 = steps_per_sec(dp_15_5)
sps_6 = steps_per_sec(dp_15_6)
sps_7 = steps_per_sec(dp_15_7)
sps_8 = steps_per_sec(dp_15_8)
sps_9 = steps_per_sec(dp_15_9)

fig,(ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9) = plt.subplots(nrows= 9,figsize = (14,10))
ax1.plot(sps_1)
ax2.plot(sps_2)
ax3.plot(sps_3)
ax4.plot(sps_4)
ax5.plot(sps_5)
ax6.plot(sps_6)
ax7.plot(sps_7)
ax8.plot(sps_8)
ax9.plot(sps_9)
ax1.set_xlim(0,45)
ax2.set_xlim(0,45)
ax3.set_xlim(0,45)
ax4.set_xlim(0,45)
ax5.set_xlim(0,45)
ax6.set_xlim(0,45)
ax7.set_xlim(0,45)
ax8.set_xlim(0,45)
ax9.set_xlim(0,45)
plt.show()

fig,ax = plt.subplots(figsize = (14,10))
ax.plot(dp_15_1+1,ls="None",marker = "|",markersize = 50,c = "black")

ax.set_ylim(0,10)
ax.set_title("Single query")
plt.show()
```
