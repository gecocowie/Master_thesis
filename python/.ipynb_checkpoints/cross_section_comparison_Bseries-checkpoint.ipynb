```python
def process_files_for_mp(file_list,common_path,save_path):
    model= "ar"
    width = 100
    jump = 10
    dim = 5
    sigma = 1
    params = [0,2]
    data_vis = "on"
    bkps_vis = "off"#"on"

    for i in range(len(file_list)):
        drifter_id = file_list[i][0:12]
        path = common_path + file_list[i]
        data, lp,bkps,lp_mean = cross_section_comp(path,drifter_id,model,width,jump,dim,sigma,params,data_vis,bkps_vis)

        saved_data = pd.DataFrame(lp_mean)        
        saved_data.to_csv(save_path+drifter_id+".csv",header = "pressure")
    return lp_mean
```
